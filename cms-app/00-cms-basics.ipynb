# The CMS package, one small step at a time

This notebook is gentle. There is no agent here and nothing clever. Each cell
does **one** small thing. You press **Shift + Enter** and read what comes out.

If you have never used a notebook: a cell is a grey box of code. Click it, press
Shift + Enter, and it runs. That is the whole skill.

Let us look at what the CMS package can do, one function at a time.

## Step 1 — log in

Run this once. It may ask for your GUC username and password.

```python
import os, getpass
from dotenv import load_dotenv

load_dotenv()  # reads your login from .env if it is there
if not os.environ.get("GUC_USERNAME"):
    os.environ["GUC_USERNAME"] = input("GUC username: ")
if not os.environ.get("GUC_PASSWORD"):
    os.environ["GUC_PASSWORD"] = getpass.getpass("GUC password: ")

from guc_cms import GucCms

cms = GucCms()
print("Logged in. ")
```

### Using GIU instead of GUC

This talks to **GUC** by default. To use **GIU** instead, pick one:

- add a line `GUC_SITE=giu` to your `.env` file, then run the cell above again, or
- change the cell above to `cms = GucCms(site="giu")`.

Your GIU username and password go in the same `GUC_USERNAME` / `GUC_PASSWORD`.
Everything below stays exactly the same: every function in this notebook is
tested end to end on both GUC and GIU.

## Step 2 — list your courses

`list_courses()` gives back every course you are registered in.

```python
courses = cms.list_courses()
print("You have", len(courses), "courses:\n")
for c in courses:
    print(" -", c.code, c.title)
```

## Step 3 — find one course

`find_course("...")` returns a single course. You can search by name or code.

```python
course = cms.find_course("Discrete Math")
print(course.code, "|", course.title)
```

## Step 4 — see the files in that course

`get_content(course)` returns the files, grouped by week.

```python
content = cms.get_content(course)
print(content.course_name, "has", len(content.items), "files.\n")
for week, files in content.weeks().items():
    print(week)
    for f in files:
        print("    ", f.kind, "-", f.title)
```

## Step 5 — get one file

`fetch_bytes(file)` downloads a file into memory. Here we just check its size.

```python
first_file = content.items[0]
data = cms.fetch_bytes(first_file)
print("Got:", first_file.title)
print("Size:", len(data), "bytes")

# To save it to a folder instead, use:
# path = cms.download(first_file, "downloads")
# print("Saved to", path)
```

## That is all

You just used the whole CMS package:

- `list_courses()` — your courses
- `find_course(name)` — one course
- `get_content(course)` — its files
- `fetch_bytes(file)` — the file itself (or `download(file, folder)` to save it)

No agent, no magic. Just five functions that read the CMS for you.
